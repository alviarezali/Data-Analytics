```python
import pandas as pd
import numpy as np
```

```python
def NewGenerates(data):
    '''Recibe una Serie de pandas donde se procedera a calcular las metricas de media y desviacion estandar para comparar
    sus valores y retornar una array que contenga los valores de:
    
    *Bajo si el valor se encuentra por debajo de la (MEDIA - DESVIACION ESTANDAR)
    *Medio si el valor se encuentra entre la [(MEDIA - DESVIACION ESTANDAR) & (MEDIA + DESVIACION ESTANDAR)]
    *Alto si el valor se encuentra por encima de la (MEDIA + DESVIACION ESTANDAR)
    
    '''
    standarDev= np.std(data)
    media = np.mean(data)
    VarCompPlus = media + standarDev
    VarCompMinus = media - standarDev
    
    conditions = [data<VarCompMinus, (data>=VarCompMinus) & (data<VarCompPlus), data>=VarCompPlus]
    options = ["Bajo", "Medio", "Alto"]
    
    return np.select(conditions, options, default="No contemplado")
```

```python
#Creamos un DateFrame con 1000 valores entre 140 y 200 con nombre de Altura en su columna
df = pd.DataFrame(data=np.random.randint(140,200,1000), columns=["Altura"])
df
```

```python
# Mostramos las metricas para que nos sirva de comparacion
print(f"Desvicion estandar {np.std(df['Altura'])}")
print(f"Media {np.mean(df['Altura'])}")
print(f"Media + std = {np.mean(df['Altura'])+np.std(df['Altura'])}")
print(f"Media - std = {np.mean(df['Altura'])-np.std(df['Altura'])}")

#LLamamos a nuestra funcion y le pasamos el DataFrame para que nos retorne nuestra nueva variable Cualitativa
df["Variable Cualitativa"]= NewGenerates(df["Altura"])
df
```
