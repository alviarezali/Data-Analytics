```python
#Importamos la librerias necesarias para el programa
import numpy as np
import pandas as pd
```

```python
#Se importan los datos con la funcion read_csv y se indica el separador del mismo
data = pd.read_csv("autos.csv", sep=";")
data.head()
```

```python
#Estudiamos los datos
#Se visualizan los valores dentro de la columna deseada

pd.unique(data["Pais_Origen"])
```

```python
#Se crea la funcion recode donde se gestionarán los datos de la columna de interes con la informacion obtenida previamente
def recode(x):
    return np.where(x=="USA",1,
                    np.where(x=="Europe",2,
                             np.where(x=="Japan",3,"Default")))
    
```

```python
# Aplicamos con el metodo apply la funcion recode 
data["Pais_Origen_num"] = data["Pais_Origen"].apply(recode)
data.sample(10)
```

```python
data.info()
```

```python
#Visualizamos con la funcion de pandas que se haya aplicado correctamente la funcion
pd.unique(data["Pais_Origen_num"])
```

```python
# Obtenemos los indices de cada uno de los paises y los almacenamos en una lista
posicion_USA = np.where(data["Pais_Origen"]=='USA')
posicion_Europe = np.where(data["Pais_Origen"]=='Europe')
posicion_Japan = np.where(data["Pais_Origen"]=='Japan')

position = [posicion_USA[0][0], posicion_Europe[0][0], posicion_Japan[0][0]]
position
```

```python
#  con el metodo loc obtenemos los primeros datos con su respectivos numeros para la columna Pais_Origen_num
data.loc[position, ["Pais_Origen", "Pais_Origen_num"]]
```
