```python
#Importamos las librerias necesarias
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
# Importamos el conjunto de datos que Utilizaremos para testear la funcion
df = pd.read_csv("autos_1.csv")
df["Si var 1 > var 2 → 5, caso contrario 3"] = df["Si var 1 > var 2 → 5, caso contrario 3"].astype(str)

df.info()
```

```python
def corr_df (df, col_country):
    '''Recibe un DataFrame y un valor en formato String de la variable a estudiar. 
    
    Esta funcion devuelve un mapa de calor con la correlacion de todas las variables numericas incluidas en el DataFrame
    
    Return: None
    '''
    
    #Agrupa los datos por Pais
    byCountry = df.groupby("Pais_Origen")
    
    #Selecciono las variables numericas del DataFrame
    num_cols = df.describe().columns.to_list()
    
    #Se crea un mapa de calor para cada pais presente.
    for country, data in byCountry:

        plt.figure(figsize=(10,5))
        corr = data[num_cols].corr()
        sns.heatmap(corr, cmap="coolwarm", annot=True, mask=np.tril(np.ones_like(corr, dtype=bool)))
        plt.tight_layout()
        plt.title(f"Tabla de correlacion para {country}")
```

```python
#LLamo a mi funcion
corr_df(df, "Pais_Origen")
```
