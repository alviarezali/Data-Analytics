```python
#Importamos la librerias necesarias para el programa
import numpy as np
import pandas as pd
```

```python
#Se importan los datos con la funcion read_csv y se indica el separador del mismo
data = pd.read_csv("autos.csv", sep=";")
data.head()
```

```python
#Observamos si hay valores Nulos
data.isna().sum()
```

```python
# Observamos valores nulos y tipos de valores
data.info()
```

```python
# Observamos las metricas del dataFrame
data.describe()
```

```python
#Obtenemos la media para las dos variables de interes

media_millas = data["Millas_Por_Galon"].mean()
media_caballos = data["Caballo_De_Fuerza"].mean()
print(media_millas)
print(media_caballos)
```

```python
#Coiamos el dataFrame en otra variable
data_copy = data.copy()
```

```python
# LLenamos los valores NaN parametricamente con la media
data_copy["Millas_Por_Galon"].fillna(media_millas, inplace=True)
data_copy["Caballo_De_Fuerza"].fillna(media_caballos, inplace=True)
```

```python
# Observamos que ya no existan valores nulos
data_copy.info()
```

```python
#Observamos que tanto han cambiado las metricas respecto al dataFrame original
data_copy.describe()
```

```python
# Obtenemos los nombres de las columnas que sean de tipo numerica


# data_name = ["Millas_Por_Galon","Cilindros","Peso_En_Libras","Desplazamiento","Aceleracion","Caballo_de_Fuerza"]
data_name = data.describe().columns.to_list()
```

```python
# Tomamos los valores del DataFrame que solo sean numericas
data_numerics = data.loc[:, data_name]
data_numerics.head()
```

```python
'''Utilizamos la libreria Sklearn para realizar una imputacion no parametrica y comparar con la imputacion parametrica realizada 
previamente'''

from sklearn.impute import KNNImputer

imputacion_NP = KNNImputer(n_neighbors=4)
data_num =imputacion_NP.fit_transform(data_numerics)
data_num
```

```python
#Convertimos en data frame los valores ya imputados
data_num= pd.DataFrame(data=data_num,
                      columns=data_name)
data_num
```

```python
#Observamos las metricas
data_num.describe()
```

```python
# Comparamos
data.describe()
```

```python
# Creamos las nuevas variables de interes
import math as ma

data_num["Potencia especifica_log10"] = (data_num["Caballo_De_Fuerza"]/data_num["Peso_En_Libras"]).apply(ma.log10)
data_num["Sqrt(var1)*exp(var2)/200"] = np.round(((data_num["Caballo_De_Fuerza"]).apply(ma.sqrt)*(data_num["Aceleracion"]).apply(np.exp))/200,2)
data_num["Si var 1 > var 2 → 5, caso contrario 3"] = np.where(data_num["Caballo_De_Fuerza"]>data_num["Peso_En_Libras"],5,3)
data_num["1/logn(var1/var2)*100"] = (1/(data_num["Caballo_De_Fuerza"]/data_num["Peso_En_Libras"]).apply(np.log))*100
data_num["var2**2/var1"] = data_num["Caballo_De_Fuerza"]**2/data_num["Peso_En_Libras"]
```

```python
# LAs visualizamos en el DataFrame
data_num.head(10)
```

```python
data_complete = pd.merge(data_copy,data_num, how="inner")
data_complete["Si var 1 > var 2 → 5, caso contrario 3"] = data_complete["Si var 1 > var 2 → 5, caso contrario 3"].astype(str)
data_complete.info()
```
